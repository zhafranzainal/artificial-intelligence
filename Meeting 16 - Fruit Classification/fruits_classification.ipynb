```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('fruits_dataset.data')
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.value_counts('fruit_name')
```

```python
features = data.iloc[:, 3:7]
label = data.iloc[:, 1]

print("Features: ")
print(features)
print("Label: ")
print(label)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=0)

print('Number of train data: ', X_train.shape[0])
print('Number of test data: ', X_test.shape[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier

# n_neighbors = number of nearest neighbors
# weights(uniform) = does not calculate distance
# weights(distance) = calculate distance
knn = KNeighborsClassifier(n_neighbors = 5,  weights="distance")

# the model takes training data consisting of features and labels
knn.fit(X_train, y_train)

# calculate model accuracy on training data
train_accuracy = knn.score(X_train, y_train)

print("The accuracy of KNN classifier on the training data is: {:.3f}".format(train_accuracy))
```

```python
# calculate model accuracy on testing data
test_accuracy = knn.score(X_test, y_test)

print("The accuracy of KNN classifier on the testing data is:  {:.3f}".format(test_accuracy))
```

```python
import pickle

# Save model that has the best accuracy
with open('knn_model.pickle', 'wb') as f:
    pickle.dump(knn, f)
```

```python
# Load model
with open('knn_model.pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
# Predict new data
new_data = [80, 4.5, 5.8, 0.77]

# Value of mass, width, height, and color_score respectively to predict the label
X = np.array([new_data])

# Carry out data prediction
prediction = model.predict(X)

print("Prediction of Species: {}".format(prediction))
```

```python
fig1 = sns.scatterplot(x='color_score', y='mass', hue='fruit_name', data=data)
fig1.set(title='Mass and Colour Score')
```

```python
fig2 = sns.scatterplot(x='height', y='width', hue='fruit_name', data=data)
fig2.set(title='Height and Width Score')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
disp.ax_.set_title("Confusion Matrix")
```
